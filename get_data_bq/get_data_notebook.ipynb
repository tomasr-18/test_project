```python


from google.cloud import bigquery
from datetime import datetime, timedelta
from dotenv import load_dotenv
import os
import pandas as pd
import db_dtypes


def fetch_data_by_company(pub_date=None, table='table_1', project_id='tomastestproject-433206', dataset='testdb_1'):
    # Ange sökvägen till din service account JSON-fil
    os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'tomastestproject-433206-adc5bc090976.json'

    table_id = f"{project_id}.{dataset}.{table}"
    # Skapa en BigQuery-klient
    client = bigquery.Client()

    # Definiera din SQL-fråga med ett parameteriserat värde för company
    # query = f"""
    #     SELECT *
    #     FROM `{table_id}`
    #     WHERE company LIKE @company_name
    # """
    query = f"""
        SELECT company,title,description,publishedAt
        FROM `{table_id}`
        WHERE LEFT(publishedAt, 10) = '{pub_date}';

        """

    query_job = client.query(query)

    # Hämta resultaten
    results = query_job.result()

    # # Skriv ut resultaten
    # for row in results:
    #     print(f"Name: {row['name']}, Age: {
    #           row['age']}, Company: {row['company']}")
    return results.to_dataframe()


# Exempel på användning av funktionen
# if __name__ == "__main__":
#     fetch_data_by_company("Google")
```
